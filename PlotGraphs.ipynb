```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt("output_TrueOnlineSarsaAgent_smallGrid.csv", delimiter=",", names=True)
```

```python
def getSplitIndices(data):
    indices=[0]
    maxidx=len(data)-1
    for idx, row in enumerate(data):
        if idx+1 <= maxidx:
            if data[idx]['NumTrainingEpisode'] > data[idx+1]['NumTrainingEpisode']:
                #[part1,part2]=np.split(data,[idx+1],axis=0)
                indices.append(idx+1)
    indices.append(len(data))
    return indices
```

```python
indices=getSplitIndices(data)
indices
```

```python
maxi=len(indices)-1
for i, idx in enumerate(indices):
    print(i,idx)
    if i+1 <= maxi:
        plt.plot(data['NumTrainingEpisode'][indices[i]:indices[i+1]], data['AvgRewardsForAll'][indices[i]:indices[i+1]])
```

```python
plt.plot(data['NumTrainingEpisode'][indices[0]:indices[1]], data['AvgRewardsForAll'][indices[0]:indices[1]])
plt.plot(data['NumTrainingEpisode'][indices[1]:indices[2]], data['AvgRewardsForAll'][indices[1]:indices[2]])
```
